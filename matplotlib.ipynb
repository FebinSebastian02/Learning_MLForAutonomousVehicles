```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(4)
y = np.arange(4)
print("x:- ", x)
print("y:- ", y)

plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,"xg") #Mark plot as green x
plt.show()
```

```python
data = np.array([1, 2, 1, 2, 5, 1, 2])
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.plot(data)
plt.show() #To render the plot in Python

```

```python
plt.title("Sample Plot")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.axis([1, 4, -2, 7]) # X axis range: (1) - (4), Y axis range: (-2) - (7)
plt.plot(data)
plt.show()
```

```python
plt.hist(data) #Plots a Histogram
plt.show()
```
