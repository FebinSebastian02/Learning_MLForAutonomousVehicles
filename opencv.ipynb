```python
import cv2 #Importing Open CV library
print(cv2.__version__)
image = cv2.imread("test_image.JPG")
print("Image type", type(image))
#cv2.imshow("Image name", image) #Shows image in a new window.
#cv2.waitKey(0) #Shows the image until a key is pressed
```

```python
import matplotlib.pyplot as plt
plt.imshow(image) 
plt.show()

# There is mismatch in colors
# Order of layers the Open CV uses is BGR whereas Matplotlib uses RGB. 
```

```python
#Therefore, we need to convert from BGR to RGB.
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
print("Image Shape", image_rgb.shape) #3D array
plt.imshow(image_rgb)
plt.show()
```

```python
#Rendering image in B&W
image_bnw = cv2.cvtColor(image_rgb,cv2.COLOR_BGR2GRAY)
print("Image Shape", image_bnw.shape) #2D array
plt.imshow(image_bnw, cmap="gray") #cmap means color map which give pixels its colors. Alternatives:- cmap = cool etc.
plt.show()
```

```python
#Re-sizing an image
image_resized = cv2.resize(image_rgb,(100,100)) #100 pixels wide and height
plt.imshow(image_resized)
plt.show()
```

```python
#Gaussian blurring using a Kernel
ksize = (55,55) #Bigger the number, bigger the smoothening/blurring. Kernel is the small matrix that slides over an image to do an operation.
image_blurred = cv2.blur(image_rgb, ksize)
plt.imshow(image_blurred)
plt.show()
```

```python
#Plotting in Subplots
fig, (ax1, ax2) = plt.subplots(2, 1, figsize = (15,10)) # 2 Rows(subplots), 1 Colomn(Figure). Default figsize is small. Therefore, enlarged. The 2 axes are returned.
ax1.imshow(image_rgb)
ax2.imshow(image_blurred)
plt.show()
```

```python
#Edge detection using Canny edge detector
image_edges = cv2.Canny(image_rgb, 125, 175) #Numerical values are the low and high thresholds for Canny edge detection algorithm.
plt.imshow(image_edges)
plt.show()
```

```python
#Plotting a rectange(Bounding box) in an image
start_point = (1500, 1500) #Top left point of rectangle. 250 from top left of image to right. 350 from that top point to down.
end_point = (3500, 3000) #Bottom right point of rectange. 
thickness = 50
box_color = (255,0,0) #Red

image_copy = image_blurred.copy() #Making a copy of the original image so that rectangles are not drawn multiple times on original image.
rectangle = cv2.rectangle(image_copy, start_point, end_point,  box_color ,thickness)
plt.imshow(rectangle)
plt.show()
```
